```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("clean_data.csv")
df.head()
```

```python
df.columns
df.shape
```

```python
print(df.columns)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=["User's ID", "ProdID", "Rating"], inplace=True)

text_columns = ["Category", "Brand", "Description", "Tags"]
df[text_columns] = df[text_columns].fillna("")

df.reset_index(drop=True, inplace=True)
```

```python
df.isnull().sum()
```

```python
df["User's ID"] = pd.to_numeric(df["User's ID"], errors='coerce')
df["ProdID"] = pd.to_numeric(df["ProdID"], errors='coerce')
df = df[(df["User's ID"] != 0) & (df["ProdID"] != 0)]
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
user_item_matrix = df.pivot_table(
    index="User's ID",
    columns="ProdID",
    values="Rating",
    fill_value=0
)
```

```python
user_item_matrix.head()
```

```python
user_item_matrix.shape
```

**The above means 1612 unique users and 1555 unique products**

```python
df.to_csv("clean_data.csv", index=False)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
user_similarity = cosine_similarity(user_item_matrix)
```
